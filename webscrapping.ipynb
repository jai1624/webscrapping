```python
#Libraries to scrap
import bs4 as bs
#Request librarie to interact with APIs
import urllib.request
```

Step1:Import above libraries 
Step2:Take the url of the website and store it a variable. 
Step3: To get the title
Step 4 : To get the url [a -represents the link for url href - represents the reference to the url]



```python
#url command for website
scr = urllib.request.urlopen('https://www.kentonline.co.uk/ashford/').read()
```

```python
#parser 
bsoup = bs.BeautifulSoup(scr,features="xml")
print(bsoup)
```

```python
#To print the title
print(bsoup.title.text)
```

```python
for link in bsoup.find_all('a'):
    print(link.get('href'))
#print(bsoup.find_all('a'))
```

```python
#To ge the title paragraph
print(bsoup.p)
```

```python
#To get the paragraph
print(bsoup.find_all('p'))
```

```python
#to get text from paragraph
print(bsoup.find('p').get_text())
```

```python
#to remove <p> tags
ptags = bsoup.find_all('p')
for  p in ptags:
    print(p.text)
```

```python
scr = urllib.request.urlopen('https://www.finder.com.au/cricket-world-cup-points-table-and-player-stats').read()
bsoup = bs.BeautifulSoup(scr,features="lxml")
tbl = bsoup.find('table')
```

```python
tbl_rows = tbl.find_all('tr')
for tr in tbl_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```
